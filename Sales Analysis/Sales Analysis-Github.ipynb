### Import Libraries 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

#### Merge 12 months of sales data rom diff libraries into one file


```python

  
folder_path = r'C:\Users\khushi shrivastava\Desktop\Datasets for Analysis\Python\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data'
file_list = glob.glob(folder_path + "\*.csv")
# file_list
Sales_Data_df = pd.DataFrame()
for i in range(0,len(file_list)):
#     print(i)
    data = pd.read_csv(file_list[i])
    df = pd.DataFrame(data)
    Sales_Data_df = pd.concat([Sales_Data_df,df],axis=0)
Sales_Data_df
```

```python
Sales_Data_df.reset_index(inplace = True, drop = True)
```

```python
Sales_Data_df
```

```python
Sales_Data_df['Month of Order']= Sales_Data_df['Order Date'].str[0:2]
```

```python
Sales_Data_df['Order Date']=Sales_Data_df['Order Date'].astype(str)
```

```python
# Sales_Data_df1['Order Date']=pd.to_datetime(Sales_Data_df1['Order Date']
```

```python
Sales_Data_df=Sales_Data_df[Sales_Data_df['Order Date'].str[0:2]!='Or']
```

```python
Sales_Data_df
```

```python
Sales_Data_df=Sales_Data_df.dropna(how='any')
Sales_Data_df
```

```python
Sales_Data_df['Price Each']=Sales_Data_df['Price Each'].astype(float)
Sales_Data_df['Quantity Ordered']=Sales_Data_df['Quantity Ordered'].astype(float)
```

```python
Sales_Data_df['Total']= Sales_Data_df['Quantity Ordered']* Sales_Data_df['Price Each']
```

```python
Sales_Data_df.reset_index(inplace = True, drop = True)
```

```python
Sales_Data_df['Month of Order']=Sales_Data_df['Order Date'].dt.month
```

```python
Sales_Data_df
```

```python
Max_Sales_Month=Sales_Data_df.groupby('Month of Order').sum()
Max_Sales_Month
```

```python
months=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', "Jul",'Aug', 'Sept', 'Oct','Nov', 'Dec']
plt.bar(months, Max_Sales_Month['Total'])
plt.xticks(months)
plt.ylabel ('Sales in USD $')
```

##### Which city had maximum Sales

```python
Sales_Data_df.head()
```

```python

def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
Sales_Data_df['Purchase City']= Sales_Data_df['Purchase Address'].apply(lambda x : f"{get_city(x)} ({get_state(x)})")
```

```python
Sales_Data_df
```

```python
Max_Sales_City=Sales_Data_df.groupby('Purchase City').sum()
```

```python
Max_Sales_City
```

```python
cities = [city for city, df in Sales_Data_df.groupby('Purchase City')]
plt.bar(cities, Max_Sales_City['Total'], color='Orange')
plt.xticks(cities, rotation='vertical', size =8)
plt.ylabel ('Sales in USD $')
```

#### What time should we advertise the product to maximise the cutomer base?

```python
Sales_Data_df['Order Date'] = pd.to_datetime(Sales_Data_df['Order Date'])
```

```python
Sales_Data_df['Hour']= Sales_Data_df['Order Date'].dt.hour
```

```python
# Max_Sales_Hour=Sales_Data_df.groupby('Hour').count()

Hour = [hr for hr, df in Sales_Data_df.groupby('Hour')]
plt.plot(Hour, Sales_Data_df.groupby(['Hour']).count(), color='Orange')
plt.xticks(Hour)
plt.ylabel ('Sales in USD $')
plt.grid()
```
